### Array indexing

```python
arr = n.array([1,2,3])
arr
```

```python
arr[1]  #Array's indexing works same as String's
```

### Array slicing

```python
arr=n.arange(2,45)  #Array's slicing also same as String's
```

```python
arr
```

```python
arr[2:19]
```

```python
arr[::]
```

```python
arr[2:34:3]
```

### NumPy Indexing and Slicing of Matrix

```python
arr = n.arange(1,31).reshape(6,5)
```

```python
arr
```

```python
slice = arr[0:2,1:3]
slice
```

```python
slice = arr[3:6,3:5]
slice
```

```python
slice = arr[2:4,1:4]
slice
arr
```

### Boolean Indexing

```python
arr = n.arange(11,21)
arr
```

```python
bool_index = arr%2==0
bool_index
```

```python
arr = arr[bool_index]
arr
```
