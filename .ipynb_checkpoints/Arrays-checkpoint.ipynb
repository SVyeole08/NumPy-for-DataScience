# Array Methods

```python
arr
```

```python
arr.min()
```

```python
arr.max()
```

```python
arr.sum()
```

```python
np.sum(arr,axis=0)  #sum of elements of matrix columnwise
```

```python
np.sum(arr,axis=1)  #sum of elements of matrix rowwise
```

```python
arr.mean()  #return mean of elements of elements matrix
```

```python
arr.std()  #return standard deviation of elements of matrix
```

```python
arr.argmax()  #return index of max element of matrix
```

```python
arr.argmin()  #return index of min element of matrix
```
